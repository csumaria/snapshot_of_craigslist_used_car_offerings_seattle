```python
list_categories=[]
for lists in listt:
    for strings in lists:
        if strings.find(':')>=0:
            new_category=strings[:strings.find(':')+1]
            if new_category not in list_categories:
                if len(new_category)<=20:
                    list_categories.append(new_category)
list_categories
```

```python
import pandas as pd
list_of_dicts=[]
for lists in listt:
    temp_dict={}
    if len(lists)>=2:
        init_str=lists[0][:4]
        year=int(init_str)
#         print(year)
        if year<2050 and year>1850:
            temp_dict['Year']=year
            temp_dict['Make Model']=lists[0][5:]
        else:
            temp_dict['Make Model']=lists[0]
#             print(year)
#             print(lists[0])
        for strings in lists:
            for attr in list_categories:
                if attr in strings:
                    new_string=strings[strings.find(':')+2:]
                    new_attr=attr[:attr.find(':')].title()
                    temp_dict[new_attr]=new_string
        if '$' in lists[len(lists)-1]:
            temp_dict['Price']=lists[len(lists)-1]
        list_of_dicts.append(temp_dict)
df1=pd.DataFrame(list_of_dicts)
```

```python
import ast

with open('make_model_3.txt', 'r') as f:
    mylist = ast.literal_eval(f.read())
```

```python
# computing dmm: a dictionary of makes and models
from collections import defaultdict
list_of_makes=[]
dict_model_makes=defaultdict(list)
for tuples in mylist:
    if tuples[1] not in list_of_makes:
        list_of_makes.append(tuples[1])

for tuples in mylist:
    dict_model_makes[tuples[1]].append(tuples[2])
# dict_model_makes
dmm={}
for k,v in dict_model_makes.items():
    vs=set(v)
    dmm[k]=list(vs)
```

```python
def Break_Make_Model_Info(string):
    sl=string.lower()
    sls=sl.split()
    for k,v in dmm.items():
        if k.lower() in sls:
            return(k)
        new=k.lower()+'_'
        if new in sls:
            return(k)
        elif 'chevy' in sls:
            return('Chevrolet')
        elif 'vw' in sls:
            return('Volkswagen')
        elif 'land' and 'rover' in sls:
            return('Land Rover')
        elif 'aston' and 'martin' in sls:
            return('Aston Martin')
        elif 'mercedes' in sls:
            return('Mercedes-Benz')
        elif 'merc' in sls:
            return('Mercedes-Benz')
```

```python
df1['Make']=df1['Make Model'].apply(Break_Make_Model_Info)
```

```python
def Break_Make_Model_Info_Diff(string):
    sl=string.lower()
    sls=sl.split()
    for k,v in dmm.items():
        if k.lower() in sls:
            sls.remove(k.lower())
        new=k.lower()+'_'
        if new in sls:
            sls.remove(new)
        elif 'chevy' in sls:
            sls.remove('chevy')
        elif 'vw' in sls:
            sls.remove('vw')
        elif 'land' in sls:
            if 'rover' in sls:
                sls.remove('land')
                sls.remove('rover')
        elif 'aston' in sls:
            if 'martin' in sls:
                sls.remove('aston')
                sls.remove('martin')
        else:
            pass
    return(sls)
```

```python
df1['Diff']=df1['Make Model'].apply(Break_Make_Model_Info_Diff)
```

```python
def from_diff_to_model(x):
    make=x[14]
    diff=x[15]
    output_string=''
    for entries in diff:
        if make!=None:
            if make=='Mercedes-Benz':
                list_mercML=['ml350','ml430','ml63','ml320','ml550']
                if entries.lower() in list_mercML:
                    return('ML-Class')
                list_mercE=['e500','e-320','e320','e350']
                if entries.lower() in list_mercE:
                    return('E-Class')
                list_mercCLK=['clk350']
                if entries.lower() in list_mercCLK:
                    return('CLK-Class')
                list_mercS=['s550','s-320','s350','s500']
                if entries.lower() in list_mercS:
                    return('S-Class')
                list_mercC=['c230','c350','300cd','c250']
                if entries.lower() in list_mercC:
                    return('C-Class')
                list_mercCLS=['cls','cla','gl320','gla','c300']
                if entries.lower() in list_mercCLS:
                    return('CLS-Class')
                list_mercSLK=['380sl','slk320']
                if entries.lower() in list_mercSLK:
                    return('SLK-Class')
            if make=='BMW':
                list_bmw3=['3-series','i3']
                if entries.lower() in list_bmw3:
                    return('3-Series')
                list_bmw4=['4-series','i4']
                if entries.lower() in list_bmw4:
                    return('4-Series')
                list_bmw5=['5-series','i5']
                if entries.lower() in list_bmw5:
                    return('5-Series')
                list_bmw6=['6-series','i6']
                if entries.lower() in list_bmw6:
                    return('6-Series')
                list_bmw7=['7-series','i7']
                if entries.lower() in list_bmw7:
                    return('7-Series')
                list_bmw8=['8-series','i8']
                if entries.lower() in list_bmw8:
                    return('8-Series')
                if type(entries)==str:
                    if entries[-1].lower()=='i':
                        num=entries[0]
                        return(num+' Series')
            for models in dmm[make]:
                modelss=models.split()
                for modelssl in modelss:
                    modelssll=modelssl.lower()
                    if entries==modelssll:
                        output_string=models
                    modelssll_=modelssll+'_'
                    if entries==modelssll_:
                        output_string=models
                    if entries=='crv':
                        output_string='CR-V'
    return(output_string)
```

```python
import re
def price_from_string(string):
    if string==(np.nan):
        return(np.nan)
    elif type(string)==str:
        ns=re.sub("[^0-9]", "", string)
        return(int(ns))
    else:
        return(string)
df1['Price']=df1['Price'].apply(price_from_string)
```
